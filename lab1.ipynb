```python
import pandas as pd
import numpy as np
aircraft = pd.read_csv("./aircraft.csv", sep=";")
airlines = pd.read_csv("./airlines.csv", sep=",")
airports = pd.read_csv("./airports.csv", sep=",")
cargo = pd.read_csv("./cargo transportation.csv", sep=";")
passenger = pd.read_csv("./passenger transportation.csv", sep=";")
```

```python
mounths = ['Январь', 'Февраль',
       'Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь',
       'Октябрь', 'Ноябрь', 'Декабрь']
mounths_year = mounths + ['Январь - Декабрь']
for m in mounths_year:
    cargo_copy = cargo[m].copy()
    cargo_copy.loc[cargo_copy == '***'] = np.NaN
    cargo_copy = cargo_copy.apply(lambda x: float(x.replace(',', '.').replace(' ', '')) if type(x) == str else x)
    cargo[m] = cargo_copy
    passenger_copy = passenger[m].copy()
    passenger_copy.loc[passenger_copy == '***'] = np.NaN
    passenger_copy = passenger_copy.apply(lambda x: float(x.replace(',', '.').replace(' ', '')) if type(x) == str else x)
    passenger[m] = passenger_copy
```

```python
cargo.info()
cargo.head()
```

```python
for name in cargo['Наименование аэропорта РФ'].unique():
    data_by_city = cargo[cargo['Наименование аэропорта РФ']==name]
    data_by_city = data_by_city[mounths_year]
    if len(data_by_city.loc[data_by_city.isnull().all(1)]) == len(data_by_city):
        cargo = cargo.drop([i for i, _ in data_by_city.iterrows()])

for name in passenger['Наименование аэропорта РФ'].unique():
    data_by_city = passenger[passenger['Наименование аэропорта РФ']==name]
    data_by_city = data_by_city[mounths_year]
    if len(data_by_city.loc[data_by_city.isnull().all(1)]) == len(data_by_city):
        passenger = passenger.drop([i for i, _ in data_by_city.iterrows()])
```

```python
for name in cargo['Наименование аэропорта РФ'].unique():
    data_by_city = cargo[cargo['Наименование аэропорта РФ']==name]
    data_by_city = data_by_city[mounths_year]
    for i, row in data_by_city.iterrows():
        if np.isnan(row['Январь - Декабрь']):
            for mounth in mounths:
                if np.isnan(row[mounth]):
                    row[mounth] = data_by_city.mean()[mounth]
            row = row.fillna(0)
            row['Январь - Декабрь'] = row[mounths].sum()
            data_by_city.loc[i] = row
    cargo.loc[cargo['Наименование аэропорта РФ']==name, mounths_year] = data_by_city
```

```python
for name in passenger['Наименование аэропорта РФ'].unique():
    data_by_city = passenger[passenger['Наименование аэропорта РФ']==name]
    data_by_city = data_by_city[mounths_year]
    for i, row in data_by_city.iterrows():
        if np.isnan(row['Январь - Декабрь']):
            for mounth in mounths:
                if np.isnan(row[mounth]):
                    row[mounth] = data_by_city.mean()[mounth]
            row = row.fillna(0)
            row['Январь - Декабрь'] = row[mounths].sum()
            data_by_city.loc[i] = row
    passenger.loc[passenger['Наименование аэропорта РФ']==name, mounths_year] = data_by_city
```

3. Построить запрос: Владелец Аэропорта, Аэропорт, Пассажиропоток суммарный за 2018, грузопоток суммарный за 2018

```python
cargo = cargo.rename(columns={'Январь - Декабрь':'Грузопоток за год'}) 
passenger = passenger.rename(columns={'Январь - Декабрь':'Пассажиропоток за год'}) 
```

```python
traffic = pd.merge(
    pd.merge(
        airports[['Наименование аэропорта РФ', 'Юрлицо']], 
        passenger[['Наименование аэропорта РФ', 'Год периода данных', 'Пассажиропоток за год']]),
    cargo[['Наименование аэропорта РФ', 'Год периода данных', 'Грузопоток за год']],
    left_on=('Наименование аэропорта РФ', 'Год периода данных'), right_on=('Наименование аэропорта РФ', 'Год периода данных')
)
traffic.loc[traffic['Год периода данных']==2018]
```

4. Перечислить аэропорты, где пассажиропоток больше медианы, а грузопоток меньше медианы


```python
passenger_cargo = pd.merge(
    cargo[['Наименование аэропорта РФ', 'Год периода данных', 'Грузопоток за год']],
    passenger[['Наименование аэропорта РФ', 'Год периода данных', 'Пассажиропоток за год']]
)
group_mean = passenger_cargo.groupby('Наименование аэропорта РФ').mean()
print(*group_mean.loc[(group_mean['Пассажиропоток за год'] > passenger['Пассажиропоток за год'].median()) & 
(group_mean['Грузопоток за год'] < cargo['Грузопоток за год'].median())].index.tolist(), sep='\n')
```

5. Перечислить авиакомпании у которых нет типов воздушных судов зарегистрированных не в России

```python
airlines = airlines[airlines['Самолеты'].notna()] #убрали отсутствующие значения
airlines
```

```python

def clean_plane_name(name): #чистка названий самолетов
    name = name.lstrip()
    ind = name.index(' ')
    return name[:ind]

notrus = list(filter(lambda row: all(clean_plane_name(plane) in russ_airplane for plane in row[1]['Самолеты'].split(',')), airlines.iterrows()))
print('Авиакомпании у которых нет типов воздушных судов зарегистрированных не в России:\n', [row[1]['Индекс'] for row in notrus])
```

6. Выведите список: Месяц, суммарный пассажиропоток за данный месяц, аэропорт, в котором пассажиропоток в данном месяце минимальный

```python
for year in passenger['Год периода данных'].unique():
    for mounth in mounths:
        print(mounth, 
              passenger.loc[passenger['Год периода данных']==year][mounth].sum(),
              passenger.loc[passenger[mounth].idxmin(), 'Наименование аэропорта РФ'],
              year
        )
```

7. Вывести список: тип аэропорта, средний грузопоток в месяц в аэропортах данного типа

```python
airports.values
```

```python
unique_types = airports['Тип'].str.strip().unique()
for airport_type in unique_types:
    airport_by_type = airports.loc[airports['Тип'].str.strip() == airport_type][['Наименование аэропорта РФ', 'Тип']]
    cargo_type = pd.merge(airport_by_type, cargo)[mounths]
    cargo_mean = cargo_type.mean().mean()
    print(airport_type, round(cargo_mean,2) if cargo_mean is not np.nan else 0)
```
